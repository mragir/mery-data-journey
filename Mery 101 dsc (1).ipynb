```python
class Student:
    def __init__(self, name,major,gpa,is_on_probation):
        self.name=name
        self.major=major
        self.gpa=gpa
        self.is_on_probation=is_on_probation
        
    def on_honor_roll(self):
        if self.gpa>=3.5:
           return True

student1=Student("jorge","bio",4.2,False)
print(student1.on_honor_roll())
```

```python
class question:
    def __init__(self,prompt,answer):
        self.prompt=prompt
        self.answer=answer
        
q_prompts=[
    "what colors are apples?\n (a) red \n (b) violeta\n (c) naranja,\n\n",
    "what color are bananas?\n (a) violeta\n (b) verdes\n (c)amarillas\n\n"
]
questions=[
    question(q_prompts[0],"a"),
    question(q_prompts[1],"c")
]

def run_test(questions):
    score=0
    for question in questions:
        answer=input(question.prompt)
        if answer==question.answer:
            score+=1
    print("you got"+str(score)+"/"+str(len(questions))+"correct")
run_test(questions)
```

```python
crops={
    "red clover":{"yield_kg":80, "usd":5.5},
    "corn":{"yield_kg":3500, "usd":0.29},
    "wheat":{"yield_kg":2000, "usd":0.19},
    "sunflower":{"yield_kg":800, "usd":1},
    "rye":{"yield_kg":1000, "usd":0.65},
}
revenues={}

for crop,data in crops.items():
    revenue=data["yield_kg"]*data["usd"]
    ars=revenue*1400
    revenues[crop]=revenue
    print(f"{crop}: yield {data['yield_kg']} kg/ha, price ${data['usd']}, revenue ${round(revenue,1)}/ha, or ${round(ars)} in ARS.")

highrevenue= max(revenues,key=lambda x:revenues[x])
print(f"\nHighest revenue: {highrevenue} with ${round(revenues[highrevenue],1)}/ha")
```
